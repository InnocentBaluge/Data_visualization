```python
import numpy as np
import pandas as pd
import folium as fo
%matplotlib.inline
```

```python
map = fo.Map()
```

```python
map
```

```python
volcano = pd.read_csv('volcano.csv')
```

```python
latitude_vol = list(volcano['Latitude'])
longitude_vol =list(volcano['Longitude'])
name_vol = list(volcano['Name'])
```

```python
vol = fo.FeatureGroup(name = 'My Map')
```

```python
for lat, lon,name in zip(latitude_vol, longitude_vol, name_vol):
    vol.add_child(fo.Marker(location =[lat,lon],
                          popup =name, icon = fo.Icon(color = 'red')))
    
```

```python
map.add_child(vol)
```

```python
vol.add_child(fo.GeoJson(data = (open('world.json','r',encoding = 'utf-8-sig').read())))
```
